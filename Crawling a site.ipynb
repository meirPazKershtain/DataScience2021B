```python
dfs = []

for i in range(700207456, 700207463, 1):
    base_url = "https://www.ficsgames.org/cgi-bin/show.cgi?ID=" + str(i)
    url = f"{base_url}"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
    response = requests.get(url, headers=headers)
    response.encoding = response.apparent_encoding
    soup = BeautifulSoup(response.text, 'html.parser')

    if response.status_code == 200:
        apartments = soup.find_all('div')
        if not apartments:
            print('er')
            break

        c = soup.find_all('script')
        d = c[1].text.strip()
        e = re.split("\'", d, 1)
        h = e[1].split("][")
        data_list = h

        # Split each element of the list into column name and value
        data_dict = {}
        for item in data_list:
            key, value = item.split(" ", 1)
            data_dict[key] = [value.strip('"')]

        # Create a DataFrame from the dictionary and append it to the list
        dfs.append(pd.DataFrame(data_dict))

# Concatenate all DataFrames in the list into a single DataFrame
df = pd.concat(dfs, ignore_index=True)

# Display the DataFrame
print(df)
```
